```python
data=pd.read_csv("/content/drive/My Drive/work/test-exam-data/code/exam_2/CancerRadiomicsKaggle_v2_050620.csv") ##Load Data
```

```python
##drop column with no character of Patient such as PatientID,Time to event,Patient Status Since we want to predict only Patient
##Status at 3 Year

data=data.drop(['Patient ID','Time to Event','Patient Status'],axis=1)
```

```python
#since the data contain  Unknown Patient Status we drop it cause unknown ground truth for status at 3 Year
data=data[data['Patient Status at 3 Year']!=-1]
```

```python
#start by define dependant variable with y and independant with x
y= data.iloc[:,6]
x=data.drop('Patient Status at 3 Year',axis=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split


X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42) #split train and test set
```

```python
# Instantiate model with 100 decision trees
rf = RandomForestClassifier(n_estimators = 1000, random_state = 42,min_samples_leaf=1, min_samples_split=2,max_depth=8)
# Train the model on training data
rf.fit(X_train, y_train)
```

```python
predictions = rf.predict(X_test) ##test_predictions on X_test of random forest
```

```python
##random forest result 

from sklearn.metrics import classification_report
from sklearn import metrics
print(metrics.confusion_matrix(y_test, predictions, labels=[0,1]))
print(classification_report(y_test, predictions))
```

```python
#now we use neuron network 
from sklearn.preprocessing import StandardScaler

# Define the scaler 
scaler = StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Import `Sequential` from `keras.models`
from keras.models import Sequential

# Import `Dense` from `keras.layers`
from keras.layers import Dense

# Initialize the constructor
model = Sequential()

# Add an input layer 
model.add(Dense(12, activation='relu', input_shape=(848,)))

# Add one hidden layer 
model.add(Dense(8, activation='relu'))

# Add an output layer 
model.add(Dense(1, activation='sigmoid'))
```

```python
# Model output shape
model.output_shape

# Model summary
model.summary()

# Model config
model.get_config()

# List all weight tensors 
model.get_weights()
```

```python
model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
                   
model.fit(X_train, y_train,epochs=20, batch_size=1, verbose=1)
```

```python
y_pred = model.predict(X_test) ##predict y_pred with neuron network model
```

```python
y_pred=(y_pred >0.05).astype(int) ##we try different treshold and choose treshold that make f1-score class 1 is at max
```

```python
from sklearn import metrics
print(metrics.confusion_matrix(y_test, y_pred, labels=[0,1]))
```

```python
#result of neuron network 

from sklearn.metrics import classification_report
target_names = ['class 0', 'class 1']
print(classification_report(y_test, y_pred, target_names=target_names))
```
